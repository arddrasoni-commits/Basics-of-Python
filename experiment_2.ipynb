```python
#CALCULATOR
a = int(input("Enter First number: "))
b = int(input("Enter Second number: "))

c = input("Enter Desired Operation (add, subtract, multiply, divide) :" )

if c == "add":
  print(a + b)
elif c == "subtract":
  print(a-b)
elif c == "multiply":
  print(a*b)
elif c == "divide":
  print(a/b)
else:
  print("invalid operation. run program again.")
```
